```python
import json
```

```python
file_path = '/content/export_37283_project-37283-at-2023-07-27-12-20-81cd22b1.json'

with open(file_path) as f:
  data = json.load(f)
```

```python
data
```

```python
dicionario = data[0]
```

```python
dicionario
```

```python
anotacoes = dicionario['annotations'][0]['result'][0]['value']
```

```python
anotacoes
```

```python
sequence = anotacoes['sequence']
```

```python
sequence
```

```python
len(sequence)
```

```python
for dic in sequence:
  x = dic['x']
  y = dic['y']
  frame = dic['frame']
  width = dic['width']
  height = dic['height']

  file_name = 'frame' + str(frame) + '.txt'
  string = '0 ' + str(x) + ' ' + str(y) + ' ' + str(width) + ' ' + str(height) + '\n' #0 é o celular

  with open(file_name, 'w') as f:
    f.writelines(string)
```
